GET/POST :   
https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en

This data returned is in specific JSON format called PxStat

Below code extracts the CSO Dataset " Exchequer Account (Historical Series)" and saves to a file **cso.json**  

```python
import requests
import json

cso_url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/"
#cso_dataset = "FIQ02"
cso_api_format = "/JSON-stat/2.0/en"

def get_cso_url(dataset):
    url = cso_url + dataset + cso_api_format
    return url

def get_cso_data(dataset:str):
    api_url = get_cso_url(dataset)
    response = requests.get(api_url)
    data = response.json()
    return data

if __name__ == "__main__":
    data = get_cso_data("FIQ02")

    with open("cso.json", "wt") as fp:
        json.dump(data, fp, indent=4)
    
    
```
