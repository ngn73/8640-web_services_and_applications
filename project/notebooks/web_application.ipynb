# Web Application   

The final end-result of the project is a graphical data-driven web application built using Flask, which integrates external APIs (Trakt and TMDb) with a MySQL database. The application dynamically renders web pages based on user requests, displaying TV show, season, and episode data that is periodically updated through scheduled background processes..

It is publicly hosted with the pythonanywhere framework : https://ngnwatchs.pythonanywhere.com/shows
