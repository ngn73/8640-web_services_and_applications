# The Movie DB (TMDB) API  
![alt text](../images/tmdb_logo.png)   
  
https://www.themoviedb.org/   
The Movie Database (TMDB) is a community built movie and TV database. Providing extensive metadata for movies, TV shows and people, they also offer one of the best selections of high resolution posters and backdrops.   

TMDB also provides an advanced API with search endpoints to extract detailed metadata on Shows/Seasons/Episodes/Cast   
Details of the TMDB APi is here ...   
https://developer.themoviedb.org/docs/getting-started  

The API is called from the baseline URL : [GET] https://api.themoviedb.org   
Example endpoints (television related) within this TMDB API include  
* Search for a show  
/3/search/tv?query=XXXXXXX
* Get show details  
/3/tv/[id]
* Get season details  
/3/tv/[id]/season/[n]
* Get cast  
/3/tv/[id]/credits  
* Get Artwork URL's  
/3/tv/[id]/images   
 
  
I used **Postman** Utility to test these endpoints extensively.

### Authenticating API Calls  

To use the TMDB API, you can request an API key by logging in to your account on TMDB, and navigating to :  
https://www.themoviedb.org/settings/api  
... to create an Application.

This will give you  
* API Read Access Token  
* API Key  

The "API Read Access Token" is used as the Bearer Token in the 'Authorization' header (curl example below).  
This Bearer Token is saved locally in a config.py file (excluded from GitHub Pushes).


```python
curl --request GET \
     --url 'https://api.themoviedb.org/3/tv/60585' \
     --header 'Authorization: Bearer <<access_token>>'
```

## Extracting a Complete TMDB Dataset  
Extracting data for over 160 shows via the TMDB API generates a significant number of requests. Since I am operating on the free tier, I implemented 1.0s delays between calls to prevent hitting rate limits.
This can result in over **4 hours** to perform a full data retrieval.

## Use of NTFY Service  

![alt text](../images/ntfy.png)

With API calls taking over 5 hours, it is impractical to be sitting at a PC reviewing logs.   
As a result, I have made use of the NTFY Service: https://ntfy.sh/ to receive notification logs directly on my Mobile while I am away from keyboard.   

This Service has it own dedicated api module in this project for handling pushing event logs to my Mobile : [api_ntfy.py](../api/api_ntfy.py)
